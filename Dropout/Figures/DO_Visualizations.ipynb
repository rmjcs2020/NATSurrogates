```python
import matplotlib.pyplot as plt
import numpy as np

params = {
    'text.latex.preamble': ['\\usepackage{gensymb}'],
    'image.origin': 'lower',
    'image.interpolation': 'nearest',
    'image.cmap': 'gray',
    'axes.grid': False,
    'savefig.dpi': 300,  # to adjust notebook inline plot size
    'axes.labelsize': 16, # fontsize for x and y labels (was 10)
    'axes.titlesize': 16,
    'font.size': 16, # was 10
    'legend.fontsize': 16, # was 10
    'xtick.labelsize': 16,
    'ytick.labelsize': 16,
    'text.usetex': True,
    'figure.figsize': [3.39, 2.10],
    'font.family': 'serif',
}
plt.rcParams.update(params)
```

```python
True_Preds = np.load('True_Preds.npy')
GP_ = np.load('GP_Preds.npy')

# Means
A_LSTM_T = np.load('A_LSTM_T_Mean.npy')
A_LSTM_T_R = np.load('A_LSTM_T_R_Mean.npy')
NA_LSTM_T = np.load('NA_LSTM_T_Mean.npy')
NA_LSTM_P = np.load('NA_LSTM_P_Mean.npy')
NA_BLSTM_P = np.load('NA_BLSTM_P_Mean.npy')
NA_TCN = np.load('NA_TCN_Mean.npy')
NA_MLP = np.load('NA_MLP_Mean.npy')
NA_MLP = NA_MLP.reshape(500,40)

# SD
A_LSTM_T_SD = np.load('A_LSTM_T_SD.npy')
A_LSTM_T_R_SD = np.load('A_LSTM_T_R_SD.npy')
NA_LSTM_T_SD = np.load('NA_LSTM_T_SD.npy')
NA_LSTM_P_SD = np.load('NA_LSTM_P_SD.npy')
NA_BLSTM_P_SD = np.load('NA_BLSTM_P_SD.npy')
NA_TCN_SD = np.load('NA_TCN_SD.npy')
NA_MLP_SD = np.load('NA_MLP_SD.npy')
NA_MLP_SD = NA_MLP_SD.reshape(500,40)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=4,figsize=(20,10))
time_gp = 0.5/499*np.arange(499)
time = 0.5/500*np.arange(500)

mode_num = 0
for i in range(2):
    for j in range(4):    
        ax[i,j].plot(time,True_Preds[:,mode_num],'bo',label='True',linewidth=3,markersize=4)        
        

ax[0,0].plot(time_gp,GP_[:,mode_num],'ro',label='GP',linewidth=3,markersize=4)

ax[0,1].plot(time,A_LSTM_T[:,mode_num],'ro',label='A-LSTM-T',linewidth=3,markersize=4)
ax[0,1].fill_between(np.arange(500)/500*0.5,A_LSTM_T[:,mode_num]+A_LSTM_T_SD[:,mode_num],A_LSTM_T[:,mode_num]-A_LSTM_T_SD[:,mode_num])


ax[0,2].plot(time,A_LSTM_T_R[:,mode_num],'ro',label='A-LSTM-T-R',linewidth=3,markersize=4)
ax[0,2].fill_between(np.arange(500)/500*0.5,A_LSTM_T_R[:,mode_num]+A_LSTM_T_R_SD[:,mode_num],A_LSTM_T_R[:,mode_num]-A_LSTM_T_R_SD[:,mode_num])

ax[0,3].plot(time,NA_TCN[0,mode_num,:],'ro',label='NA-TCN',linewidth=3,markersize=4)
ax[0,3].fill_between(np.arange(500)/500*0.5,NA_TCN[0,mode_num,:]+NA_TCN_SD[mode_num,:],NA_TCN[0,mode_num,:]-NA_TCN_SD[mode_num,:])

ax[1,0].plot(time,NA_MLP[:,mode_num],'ro',label='NA-MLP',linewidth=3,markersize=4)
ax[1,0].fill_between(np.arange(500)/500*0.5,NA_MLP[:,mode_num]+NA_MLP_SD[:,mode_num],NA_MLP[:,mode_num]-NA_MLP_SD[:,mode_num])


ax[1,1].plot(time,NA_LSTM_T[:,mode_num],'ro',label='NA-LSTM-T',linewidth=3,markersize=4)
ax[1,1].fill_between(np.arange(500)/500*0.5,NA_LSTM_T[:,mode_num]+NA_LSTM_T_SD[:,mode_num],NA_LSTM_T[:,mode_num]-NA_LSTM_T_SD[:,mode_num])

ax[1,2].plot(time,NA_LSTM_P[0,mode_num,:],'ro',label='NAT-LSTM-PT',linewidth=3,markersize=4)
ax[1,2].fill_between(np.arange(500)/500*0.5,NA_LSTM_P[0,mode_num,:]+NA_LSTM_P_SD[mode_num,:],NA_LSTM_P[0,mode_num,:]-NA_LSTM_P_SD[mode_num,:])

ax[1,3].plot(time,NA_BLSTM_P[0,mode_num,:],'ro',label='NAT-BLSTM-PT',linewidth=3,markersize=4)
ax[1,3].fill_between(np.arange(500)/500*0.5,NA_BLSTM_P[0,mode_num,:]+NA_BLSTM_P_SD[mode_num,:],NA_BLSTM_P[0,mode_num,:]-NA_BLSTM_P_SD[mode_num,:])


for i in range(2):
    for j in range(4):
        ax[i,j].legend(loc='lower right',fontsize=18)


plt.tight_layout()
plt.savefig('Figure_3a.png')
plt.show()
```

```python
# Plot testing rmses
gp_error = np.mean((True_Preds[:499,:]-GP_[:,:])**2)
print('GP error:',gp_error)

a_lstm_t_error = np.mean((True_Preds[:,:]-A_LSTM_T[:,:-2])**2)
print('A-LSTM-T error:',a_lstm_t_error)

a_lstm_t_r_error = np.mean((True_Preds[:,:]-A_LSTM_T_R[:,:-2])**2)
print('A-LSTM-T-R error:',a_lstm_t_r_error)

na_tcn_error = np.mean((True_Preds[:,:]-np.transpose(NA_TCN[0,:-2,:]))**2)
print('NAT TCN error:',na_tcn_error)

na_mlp_error = np.mean((True_Preds[:,:]-NA_MLP[:,:])**2)
print('NAT MLP error:',na_mlp_error)

na_lstm_t_error = np.mean((True_Preds[:,:]-NA_LSTM_T[:,:])**2)
print('NAT LSTM TP error:',na_lstm_t_error)

na_lstm_p_error = np.mean((True_Preds[:,:]-np.transpose(NA_LSTM_P[0,:-2,:]))**2)
print('NAT LSTM PT error:',na_lstm_p_error)

na_blstm_p_error = np.mean((True_Preds[:,:]-np.transpose(NA_BLSTM_P[0,:-2,:]))**2)
print('NAT BLSTM PT error:',na_blstm_p_error)
```
