```python
import matplotlib.pyplot as plt
import numpy as np

params = {
    'text.latex.preamble': ['\\usepackage{gensymb}'],
    'image.origin': 'lower',
    'image.interpolation': 'nearest',
    'image.cmap': 'gray',
    'axes.grid': False,
    'savefig.dpi': 300,  # to adjust notebook inline plot size
    'axes.labelsize': 16, # fontsize for x and y labels (was 10)
    'axes.titlesize': 16,
    'font.size': 16, # was 10
    'legend.fontsize': 16, # was 10
    'xtick.labelsize': 16,
    'ytick.labelsize': 16,
    'text.usetex': True,
    'figure.figsize': [3.39, 2.10],
    'font.family': 'serif',
}
plt.rcParams.update(params)
```

```python
True_Preds = np.load('True_Preds.npy')
GP_ = np.load('GP_Preds.npy')
A_LSTM_T = np.load('A_LSTM_T_Preds.npy')
A_LSTM_T_R = np.load('A_LSTM_T_R_Preds.npy')
NA_LSTM_T = np.load('NA_LSTM_T_Preds.npy')
NA_LSTM_P = np.load('NA_LSTM_P_Preds.npy')
NA_BLSTM_P = np.load('NA_BLSTM_P_Preds.npy')
NA_TCN = np.load('NA_TCN_Preds.npy')
NA_MLP = np.load('NA_MLP_Preds.npy')
```

```python
# Reshape MLP
NA_MLP = NA_MLP.reshape(500,40)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=4,figsize=(20,10))
time_gp = 0.5/499*np.arange(499)
time = 0.5/500*np.arange(500)

mode_num = 0
for i in range(2):
    for j in range(4):    
        ax[i,j].plot(time,True_Preds[:,mode_num],label='True',linewidth=3)        
        
        
ax[0,0].plot(time_gp,GP_[:,mode_num],label='GP',linewidth=3)
ax[0,1].plot(time,A_LSTM_T[:,mode_num],label='A-LSTM-T',linewidth=3)
ax[0,2].plot(time,A_LSTM_T_R[:,mode_num],label='A-LSTM-T-R',linewidth=3)
ax[0,3].plot(time,NA_TCN[0,mode_num,:],label='NA-TCN',linewidth=3)

ax[1,0].plot(time,NA_MLP[:,mode_num],label='NA-MLP',linewidth=3)
ax[1,1].plot(time,NA_LSTM_T[0,:,mode_num],label='NA-LSTM-T',linewidth=3)
ax[1,2].plot(time,NA_LSTM_P[0,mode_num,:],label='NA-LSTM-P',linewidth=3)
ax[1,3].plot(time,NA_BLSTM_P[0,mode_num,:],label='NA-BLSTM-P',linewidth=3)


for i in range(2):
    for j in range(4):
        ax[i,j].legend(loc='lower right',fontsize=20)
        ax[i,j].set_ylim(0.0,0.5)


plt.tight_layout()
plt.savefig('Figure_1.png')
plt.show()
```

```python
print('GP Error:',np.mean((True_Preds[:499,:]-GP_[:,:])**2))
print('A_LSTM_T Error:',np.mean((True_Preds[:,:]-A_LSTM_T[:,:-2])**2))
print('A_LSTM_T_R Error:',np.mean((True_Preds[:,:]-A_LSTM_T_R[:,:-2])**2))
print('NA_TCN Error:',np.mean((True_Preds[:,:]-np.transpose(NA_TCN[0,:-2,:]))**2))
print('NA_MLP Error:',np.mean((True_Preds[:,:]-NA_MLP[:,:])**2))
print('NA_LSTM_T Error:',np.mean((True_Preds[:,:]-NA_LSTM_T[0,:,:])**2))
print('NA_LSTM_P Error:',np.mean((True_Preds[:,:]-np.transpose(NA_LSTM_P[0,:-2,:]))**2))
print('NA_BLSTM_P Error:',np.mean((True_Preds[:,:]-np.transpose(NA_BLSTM_P[0,:-2,:]))**2))
```
